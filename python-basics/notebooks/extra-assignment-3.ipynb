```python
# print 'CORRECT' if i == 10
num1 = int(input("Enter any number which is divisible by 10:"))
result = num1 % 10
if result == 0:
    print("Correct")
else:
    print("Not Correct")
```

```python
# Catagory the people by their age like children, adult, citizen, senior citizen...
age = int(input("Enter your age: "))

if age < 0:
    print("Invalid age!")
elif age <= 12:
    print("Children")
elif age <= 20:
    print("Adult")
elif age <= 59:
    print("Citizen")
else:
    print("Senior Citizen")
```
