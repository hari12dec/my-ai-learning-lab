```python
# print 0 to 20 by using range
for temp in range(0, 21):
    print(temp)
```

```python
# print range 10 to 20
for temp in range(10, 20):
    print(temp)
```

```python
# print range 1 to 20 with step 2
for temp in range (1, 20, 2):
    print(temp)
```

```python
# Print number of items in the list by using 'len'
list = input("Enter number separated by single white space").split()
print(f"Length of the list {len(list)}")
```
