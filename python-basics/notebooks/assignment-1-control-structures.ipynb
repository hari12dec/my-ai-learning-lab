```python
# Body Mass Index

# BMI Weight Status
# Below 18.5 -- Underweight
# 18.5 - 25.9 -- Healthy Weight
# 26.0 - 29.9 -- Overweight
# 30.0+ -- Obesity
# 40.0+ -- Severe obesity
```

```python
bmi = float(input("Enter the BMI Index:"))

if bmi < 18.5:
    print("Underweight")
elif bmi <= 25.9:
    print("Healthy Weight")
elif bmi <= 29.9:
    print("Overweight")
elif bmi <= 39.9:
    print("Obesity")
else:
    print("Severe Obesity")
```
