```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("CKD.csv")
independent = dataset[[
  'age', 'bp', 'sg', 'al', 'su', 'rbc', 'pc', 'pcc', 'ba',
  'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv', 'wc',
  'rc', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane'
]]
dependent = dataset['classification']
```

```python
independent = pd.get_dummies(independent, drop_first = True)
independent
```

```python
columns = independent.columns.tolist()
columns
```

```python
dependent.value_counts(normalize=True) * 100
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 1/3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV

# Instantiate the model
gnb = GaussianNB()

# Hyperparameter grid
param_grid = {
    'var_smoothing': [1e-09, 1e-08, 1e-07, 1e-06]  # typical values to try
}

# GridSearchCV
grid = GridSearchCV(
    estimator=gnb,
    param_grid=param_grid,
    scoring='recall',  # prioritize minimizing false negatives
    cv=5,
    n_jobs=-1
)

# Fit the model
grid.fit(X_train, y_train)

# Best parameters
print("Best Params:", grid.best_params_)

# Predictions
y_pred = grid.predict(X_test)
```

```python
# fitting the model for grid search 
grid.fit(X_train, y_train) 
```

```python
# print best parameter after tuning 
print(grid.best_params_) 
re=grid.cv_results_
# print(re)

from sklearn.metrics import confusion_matrix
grid_predictions = grid.predict(X_test) 
cm = confusion_matrix(y_test, grid_predictions)
print(cm)

# print classification report 
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)

print(clf_report)
```

```python
from sklearn.metrics import roc_auc_score

# roc_auc_score(y_test,grid.predict_proba(X_test)[:,1])

y_prob_yes = grid.best_estimator_.predict_proba(X_test)[:, 1]
roc_auc = roc_auc_score(y_test, y_prob_yes)
print(roc_auc)
```
