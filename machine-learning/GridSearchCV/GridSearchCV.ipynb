```python
import pandas as pd
insuranceDataSets = pd.read_csv("insurance_pre.csv") # reading csv and load to variable
```

```python
insuranceDataSets = pd.get_dummies(insuranceDataSets, dtype = int, drop_first=True) # categorical column to value
insuranceDataSets.columns
#spliting input and output 
independent = insuranceDataSets[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent = insuranceDataSets[['charges']]
```

```python
#spliting test data and train data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 1/3, random_state = 0)

# Convert y to 1D
y_train = y_train.values.ravel()
y_test = y_test.values.ravel()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR

param_grid = {'kernel':['rbf', 'poly', 'linear', 'sigmoid'], 'C':[10, 100, 1000, 3000], 'gamma': ['auto', 'scale']} # creating possible param for model creation
grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 0, n_jobs = -1)

grid.fit(X_train, y_train)
```

```python
result = grid.cv_results_
print(f"best parameter {grid.best_params_}")
```

```python
table = pd.DataFrame.from_dict(result)
table
```
