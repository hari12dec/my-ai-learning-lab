```python
import pandas as pd
startupDataSet = pd.read_csv("50_Startups.csv") # reading csv and load to variable
```

```python
startupDataSet = pd.get_dummies(startupDataSet, dtype = int, drop_first=True) # categorical column to value
#spliting input and output 
independent = startupDataSet[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
dependent = startupDataSet[['Profit']]
```

```python
#spliting test data and train data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 1/3, random_state = 0)

# Convert y to 1D
y_train = y_train.values.ravel()
y_test = y_test.values.ravel()
```

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Create a pipeline with scaling + model
model = SVR(kernel='sigmoid', C=10000.0)

# Fit the model
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_pred)
r_score
```

```python
import pickle
pickle.dump(model, open("svr.sav", "wb"))
```
