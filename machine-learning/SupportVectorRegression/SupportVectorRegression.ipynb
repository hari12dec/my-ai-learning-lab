```python
import pandas as pd
startupDataSet = pd.read_csv("50_Startups.csv") # reading csv and load to variable
```

```python
startupDataSet = pd.get_dummies(startupDataSet, dtype = int, drop_first=True) # categorical column to value
#spliting input and output 
independent = startupDataSet[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
dependent = startupDataSet[['Profit']]
```

```python
#spliting test data and train data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(independent, dependent, test_size = 0.4, random_state = False)
```

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# Create a pipeline with scaling + model
model = Pipeline([
    ('scaler', StandardScaler()),
    ('svr', SVR(kernel='linear', C=0.1))
])

# Fit the model
model.fit(X_train, Y_train)

# Predict
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_predict, Y_test)
```

```python
import pickle
pickle.dump(model, open("svr.sav", "wb"))
```
